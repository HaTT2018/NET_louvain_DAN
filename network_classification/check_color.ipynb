```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 第一张图，假如两条曲线（蓝色和橙色）重合，证明revised数据（来自J.L.，无坐标）和源数据（来自K.Z.，有坐标）是同一个数据
# 第二张图代表了这个数据的空间位置

```python
color = 'b'

data = pd.read_csv('../data/revised_data/flow_'+color+'_revised.csv')
data_ref = pd.read_csv('../data/revised_data/reference/'+color+'_q_reg_40.csv', index_col=0)
plt.figure(figsize=[16,4])
data.iloc[0,:].plot()
data_ref.iloc[:, 4:-1].iloc[0,:].plot()

plt.figure(figsize=[10,10])
plt.plot(data_ref.iloc[:, 2], data_ref.iloc[:, 3], '.', color=color)
plt.xlim([116.25, 116.5])
plt.ylim([39.8, 40.025])
```
