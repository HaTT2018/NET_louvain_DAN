```python
res = 11
seed = 25

id_402 = pd.read_csv('../res/%i_res%i_id_402_withclass.csv'%(seed, res), index_col=0)
b_det = pd.read_csv('../res/%i_res%i_b_det.csv'%(seed, res), index_col=0)
part1 = pd.read_csv('../res/%i_res%i_det_partition_results1.csv'%(seed, res))
part2 = pd.read_csv('../res/%i_res%i_det_partition_results2.csv'%(seed, res))
color_set = ['blue', 'green', 'orange', 'black', 'red']

# this code (for loop) shows id_402 has experienced boundary adjustment
for i in range(len(id_402)):
    det_id = id_402.loc[i, 'id']
    cls_402 = id_402.loc[i, 'class_i']
    try:
        cls_det = part1[part1['det'] == det_id]['0'].values[0]
        if cls_402 != cls_det:
            part1.loc[part1['det'] == det_id, '0'] = cls_402
            print(i)
    except:
        cls_det = part2[part2['det'] == det_id]['0'].values[0]
        if cls_402 != cls_det:
            part2.loc[part2['det'] == det_id, '0'] = cls_402
            print(i)
    


fig = plt.figure(figsize=[15,15], dpi=125)
ax1 = fig.add_subplot(111)
for i in range(len(id_402)):
    det_id = id_402.loc[i, 'id']
    x = id_402.loc[i, 'lat']
    y = id_402.loc[i, 'long']
    try:
        cls = part1[part1['det'] == det_id]['0'].values[0]
        ax1.plot(x, y, color=color_set[cls], marker='+', markersize=10)
    except:
        cls = part2[part2['det'] == det_id]['0'].values[0]
        ax1.plot(x, y, color=color_set[cls], marker='o', markersize=10, markerfacecolor='none')
```
