```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import ipdb
```

```python
dir_ = os.listdir('../res/data_rel_class_NSk/')
```

```python
res_set = []

for i in dir_:
    if i.split('.')[-1] != 'csv':
        continue
    res_term = i.split('_')[1]
    res_ind = int(res_term[res_term.find('s')+1:])
    if not (res_ind in res_set):
        res_set.append(res_ind)
```

```python
rel_table = pd.DataFrame([], columns=['randseed', 'resolution', 'num_class', 'NSk_mean', 'NSk_min', 'NSk_max', 'NSk_std', 'TV', 'CH'])
```

```python
for i in range(len(dir_)):
    try:
        index_ind = dir_[i].split('_')[2]
    except:
        continue
        
    if index_ind == 'index.csv':
        res_term = dir_[i].split('_')[1]
        res_ind = int(res_term[res_term.find('s')+1:])
        
        data = pd.read_csv('../res/data_rel_class_NSk/'+dir_[i])
        rel_table.loc[i, 'resolution'] = res_ind
        rel_table.loc[i, ['NSk_mean', 'NSk_min', 'NSk_max', 'NSk_std', 'TV', 'CH']] = data.loc[0, ['NSk_mean', 'NSk_min', 'NSk_max', 'NSk_std', 'TV', 'CH']]
        
        data2 = pd.read_csv('../res/data_rel_class_NSk/'+dir_[i+1])
        rel_table.loc[i, 'num_class'] = data2.iloc[:, 0].drop_duplicates().shape[0]
        
        rel_table.loc[i, 'randseed'] = int(dir_[i].split('_')[0])
        
```

```python
rel_table.loc[(rel_table['num_class']==5) & (rel_table['NSk_mean']<0.89)].sort_values(by='CH')
```

```python
class_set = list(rel_table['num_class'].drop_duplicates().sort_values().values)
class_NSk_min = [rel_table[rel_table['num_class']==i]['NSk_mean'].values.min() for i in class_set]
class_TV_min = [rel_table[rel_table['num_class']==i]['TV'].values.min() for i in class_set]
class_CH_min = [rel_table[rel_table['num_class']==i]['CH'].values.min() for i in class_set]

fig = plt.figure(figsize=[15, 4], dpi=100)
ax1 = fig.add_subplot(131)
ax1.plot(class_set, class_NSk_min)
ax1.plot(class_set, class_NSk_min, 'ro')
ax1.set_xlabel('number of classes')
ax1.set_ylabel('NSk mean')

ax2 = fig.add_subplot(132)
ax2.plot(class_set, class_TV_min)
ax2.plot(class_set, class_TV_min, 'ro')
ax2.set_xlabel('number of classes')
ax2.set_ylabel('TV mean')

ax1 = fig.add_subplot(133)
ax1.plot(class_set, class_CH_min)
ax1.plot(class_set, class_CH_min, 'ro')
ax1.set_xlabel('number of classes')
ax1.set_ylabel('CH mean')
```
