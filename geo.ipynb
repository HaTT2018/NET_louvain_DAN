```python
def select_class_id (t, lat_list_all):
    lat_list_all_lat = lat_list_all.lat

    lat_list_class_i = id0[id0['class'] == t].lat
    id_selected = []
    for i in range(len(lat_list_all_lat)):
        if lat_list_all_lat[i] in list(lat_list_class_i):
            id_selected.append(lat_list_all.id[i])
    return id_selected
```
